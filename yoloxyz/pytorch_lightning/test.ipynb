```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
name = "00000324"

# ng dn n nh v file label
image_path = "C:/Users/admin/Desktop/datasets/train/images/"+ name + ".jpg"  # Thay th bng ng dn thc t
label_path = "C:/Users/admin/Desktop/datasets/train/labels/"+ name + ".txt"  # Thay th bng ng dn thc t

# c nh
img = plt.imread(image_path)

# c file label
with open(label_path, 'r') as f:
    labels = []
    for line in f:
        label = list(map(float, line.strip().split()))
        labels.append(label)

# Chuyn i labels sang numpy array
labels = np.array(labels)

# To figure v axes
fig, ax = plt.subplots(1)

# Hin th nh
ax.imshow(img)

# V bounding boxes
for label in labels:
    class_id = int(label[0])
    x_center = label[1] * img.shape[1]
    y_center = label[2] * img.shape[0]
    width = label[3] * img.shape[1]
    height = label[4] * img.shape[0]

    x1 = x_center - width / 2
    y1 = y_center - height / 2

    # To rectangle patch
    rect = patches.Rectangle((x1, y1), width, height, linewidth=1, edgecolor='r', facecolor='none')

    # Thm rectangle vo axes
    ax.add_patch(rect)
    ax.text(x1, y1, str(class_id), color='r', fontsize=10)


# Hin th plot
plt.show()
```
